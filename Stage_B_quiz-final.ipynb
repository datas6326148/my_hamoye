# Non Normalized Version

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('energydata_complete.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum(axis = 0)
```

```python
X = df['T2']
y = df['T6']
```

```python
X=df['T2'].values.reshape(-1,1)
y=df['T6'].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=77, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
y_predict= lin_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_predict))
round(rmse,3)
```

```python
df = df.drop(columns=['date','lights'],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
normalized_df.head()
```

```python
feat_df = df.drop(columns=['Appliances'])
appliances = df['Appliances']
```

```python
from sklearn.model_selection import train_test_split
lin_reg = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(feat_df, appliances, random_state=42, test_size=0.3)
lin_reg.fit(X_train,y_train)
y_predict= lin_reg.predict(X_test)
```

```python
y_predict_train=lin_reg.predict(X_train)
from sklearn.metrics import mean_absolute_error
mae_train = mean_absolute_error(y_train, y_predict_train)
print('mae_train: ',round(mae_train,3))
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_predict)
print('mae_test: ',round(mae,3))
```

```python
rmse_feat_train = np.sqrt(mean_squared_error(y_train, y_predict_train))
print('rmse_train: ',round(rmse_feat_train,3))
```

```python
from sklearn.metrics import mean_squared_error
rmse_feat = np.sqrt(mean_squared_error(y_test, y_predict))
print('rmse_test: ',round(rmse_feat,3))
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train,y_train)
y_predict_ridge = ridge.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse_feat_ridge = np.sqrt(mean_squared_error(y_test, y_predict_ridge))
round(rmse_feat_ridge,3)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train,y_train)
y_predict_lasso = lasso.predict(X_test)
```

```python
def get_weights_df(model, feat, col_name):
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

lasso_model_weights = get_weights_df(lasso, X_train, 'Lasso_Model_Weight')
lasso_model_weights
```

```python
rmse_lasso = np.sqrt(mean_squared_error(y_test,y_predict_lasso))
round(rmse_lasso,3)
```
